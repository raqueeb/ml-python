```
# ধন্যবাদ Josh Mancusoকে 
# http://bit.ly/gapminder-test

import numpy as np
import pandas as pd
import plotly_express as px

#colab notebook plotly express display function:

def configure_plotly_browser_state():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              plotly: 'https://cdn.plot.ly/plotly-latest.min.js?noext',
            },
          });
        </script>
        '''))

#Import and create the dataframes:

income = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--datapoints--income_per_person_gdppercapita_ppp_inflation_adjusted--by--geo--time.csv')
lifespan = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--datapoints--life_expectancy_years--by--geo--time.csv')
population = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--datapoints--population_total--by--geo--time.csv')
entities = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv')

#Merge them together into one DF:
df = pd.merge(income, lifespan).merge(population)
df = pd.merge(df, entities[['country', 'name', 'world_6region']], how='inner', left_on='geo', right_on='country')

#Drop some columns we don't need and rename columns:
df = df.drop(columns=['geo', 'country'])
df = df.rename(columns={
    'time': 'year',
    'life_expectancy_years':'lifespan',
    'income_per_person_gdppercapita_ppp_inflation_adjusted': 'income',
    'population_total':'population',
    'name':'country',
    'world_6region':'region'
})


```

```
#First lets do a scatter plot for 2018:
configure_plotly_browser_state()
px.scatter(df[df.year==2018], x="income", y="lifespan", color="region", size="population", log_x=True, size_max=60, hover_name="country")
```

```
#Now lets try animated version:

configure_plotly_browser_state()
px.scatter(df, x="income", y="lifespan",size="population", size_max=60, color="region", hover_name="country", 
           animation_frame="year", animation_group="country", log_x=True, range_x=[300,100000], range_y=[20,90],
            labels=dict(population="Population", income="GDP per Capita", lifespan="Life Expectancy"))
```
